### ***INPLACE***

```python
## Let's Operate on some missing data
## Pandas describe missing data as NaN
car_sales_missing = pd.read_csv("car-sales-missing-data.csv")
car_sales_missing
```

```python
## To fill the missing data
## Let's fill the missing data with the mean of Odometer column
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
```

```python
car_sales_missing.head() # We still have Nan, we will have to use INPLACE
```

```python
## INPLACE
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean(),
                                     inplace = True)

car_sales_missing
```

### ***DROPNA***

```python
## DROPNA will drop the tuples which have any attribute as nan
car_sales_missing.dropna(inplace = True)
car_sales_missing
```

## **CREATING NEW COLUMNS**

```python
# Adding new columns in car sales
# Column from Series
seats_column = pd.Series([5, 5, 5, 5, 5])

# New Columns called seats
car_sales["Seats"] = seats_column
car_sales
```

```python
## Fixing the Nan Seats value
car_sales["Seats"].fillna(5, inplace = True)
car_sales.tail()
```

```python
## Creating a column from Python lists
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 4.7, 3.8, 3.0, 6.5, 9.0]
car_sales["Fuel per 100KM"] = fuel_economy
car_sales

# A list has to be same length of dataframe, whereas in series it can be smaller
# and the rest will be NanN. But lists which are smaller than dataframe will
# give error
```

### ***CREATING A COLUMN FROM ANOTHER COLUMN***
`TEST_CASE - ` We need to see how many litres this car has used in its life

```python
car_sales["Total Fuel Used"] = car_sales["Odometer (KM)"] / 100 * car_sales["Fuel per 100KM"]
car_sales
```

```python
## Create a column of single values
car_sales["No Of Wheels"] = 4
car_sales.head(3)
```

```python
## Another Column
car_sales["Passed Road Safety"] = True
car_sales.dtypes
```

## **REMOVING A COLUMN**

```python
## Removing passed road safety
car_sales.drop(["Passed Road Safety"], axis = 1, inplace = True)
car_sales
```

## **TAKING A SAMPLE OF DATA (SOME PART)**

```python
car_sales.sample(frac=0.5) # Takes 50% and returns shuffled
```

```python
# Only select 20% of data
car_sales.sample(frac=0.2)
```

```python
car_sales.sample(frac=1) # Takes whole and returns shuffled
```

```python
car_sales_shuffled = car_sales.sample(frac=1)
car_sales_shuffled  # We get shuffled data
```

```python
## Arranging indexes back in order
car_sales_shuffled.reset_index(inplace = True, drop=True)
```
