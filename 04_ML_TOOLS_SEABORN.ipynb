# ***SEABORN***

```python
import seaborn as sns
%matplotlib inline
```

```python
# LOADING A SEABORN DEFAULT DATAFRAME
tips = sns.load_dataset('tips')
tips.head()
```

```python
len(tips)
```

## **1. DISTRIBUTION (DIST) PLOT**

```python
## DISTPLOT - PASS SINGLE COLUMN
sns.distplot(tips["total_bill"]) # see how distributed the total bill data is
```

```python
# REMOVE THE LINE CALLED KDE
sns.distplot(tips["total_bill"], kde = False)
```

```python
# SET BINS
sns.distplot(tips["total_bill"], kde = False, bins=30)
```

## **2. JOINT PLOTS - Match 2 DISTPLOTS**

```python
## JOINTPLOTS ARE USED TO JOIN 2 DISTPLOTS
sns.jointplot(x="total_bill" ,y="tip", data=tips)
```

```python
# Kind attribute
sns.jointplot(x="total_bill",y="tip",data=tips,kind="hex")
```

```python
# linear regression line
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg")
```

```python
# kde -> where the datapoints are most concentrated
sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde")
```

## **3.PAIRPLOT - PLOT PAIRWISE RELATIONSHIPS ACROSS ENTIRE DATAFRAME**

```python
## PAIRPLOT --> DOEST THE JOINTPLOT FOR ALL SINGLE NUMERICAL COMBINAAITONS IN DATAFRAME
sns.pairplot(tips, hue="sex") # hue -> color the datapoints of the hue
```

## **4. RUGPLOTS**

```python
## RUGPLOTS -->
sns.rugplot(tips['total_bill'])
```

# ***CATEGORICAL PLOTS***

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

### **BARPLOT - CATEGORICAL**

```python
## SIMPLE BARPLOT
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
## ESTIMATOR
import numpy as np
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

### **COUNTPLOT**

```python
## COUNTPLOT --> Counts the number of occurences
sns.countplot(x='sex', data=tips)
```

### **BOXPLOT**

```python
## BOXPLOT
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
# Compare total bill daybyday and also if the person is smoker or not
```

### **STRIPPLOT**

```python
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True, hue="sex")
```

# ***MATRIX PLOTS***

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

### **CORRELATIOND DATA AND HEATMAP**

```python
tips_numerical = tips.select_dtypes(['number'])
tips_numerical
```

```python
tips_numerical.corr()
```

```python
sns.heatmap(tips_numerical.corr(),
            annot=True)
```

```python
sns.heatmap(tips_numerical.corr(),
            annot=True,
            cmap='coolwarm')
```

```python
flights.head()
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(fp,
            linecolor='white',
            linewidth=3)
```

### **CLUSTERMAP**

```python
sns.clustermap(fp)
```
