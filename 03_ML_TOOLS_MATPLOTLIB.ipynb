# ***MATPLOTLIB***

## **SIMPLE PLOTS**

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## CREATING A PLOT WITH 2 NUMPY ARRAYS
import numpy as np
x = np.linspace(0, 5, 11)
y = x**2
x
```

```python
## PLOTTING WAY 1 - Functional
plt.plot(x, y)
plt.show()
```

```python
## ADDING X label, Y label and TITLE
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('TiTlE')
plt.show() ## Show the plot
```

```python
## PUTTING MULTIPLE PLOTS ON THE SAME CAMPUS
plt.subplot(1, 2, 1) # no. of rows, np. of columns, plot no. we are referring to
plt.plot(x, y, color='r')

plt.subplot(1, 2, 2) # no. of rows, np. of columns, plot no. we are referring to
plt.plot(y, x, color='b')

plt.show()
```

```python
## OBJECT ORIENTED METHODS
fig = plt.figure() # creates a figure object

# [left, bottom, width, height] ranging from 0-1 (percent of blank canvas we need to take)
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # CREATES A FIGURE OF FOLLOWING DIMENSIONS

# Plot on this axes
axes.plot(x, y)
axes.set_xlabel("XLABEL")
axes.set_ylabel("YLABEL")
axes.set_title("TITTTLLLEEE")
```

```python
## Putting 2 sets of figures on one canvas
fig = plt.figure() # Create the figure object

# [left, bottom, width, height] ranging from 0-1 (percent of blank canvas we need to take)
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # 20% from the left, 50% of bottom, 40% of width and 30% height

axes1.plot(x,y)
axes2.plot(y,y)

axes1.set_xlabel("XLABEL")
axes1.set_ylabel("YLABEL")
axes1.set_title("Bigger Plot")

axes2.set_xlabel("XLABEL")
axes2.set_ylabel("YLABEL")
axes2.set_title("Larger Plot")
```

```python
## CREATING SUBPLOTS USING OBJECT ORIENTED PROGRAMMING
fig, axes = plt.subplots(nrows=1, ncols=2)
# Gives 2 columns
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3)
plt.tight_layout() # resolve overlap issues
```

```python
## SUBPLOTS USING OBJECT ORIENTED PROGRAMMING
fig,axes = plt.subplots(nrows=1, ncols=2)

# axes = list of axeses
for current_axes in axes:
  current_axes.plot(x, y)
```

```python
## WE CAN ALSO ACCESS USING INDEXING
fig,axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, x)
axes[0].set_title("AXES 1")

axes[1].plot(y, x)
axes[1].set_title("AXES 2")
```

## **FIGURE SIZE AND DPI**

```python
fig = plt.figure(figsize=(10, 2), dpi=100)
# figsize = plot
# dpi = pixels per inch

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y)
```

```python
## ALSO DOING THIS WITH SUBPLOTS
fig, axes = plt.subplots(nrows = 1, ncols = 5,figsize=(5, 6), dpi=120)

for i in axes:
  i.plot(x, x)

plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 4), dpi=110)
axes[0].plot(x, x)
axes[1].plot(x, x**4)
```

## **PLOT APPEARANCE**

#### **COLORS**

```python
# CREATING A PLOT
fig = plt.figure(figsize=(3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color='y') # Changing color to yellow
```

```python
fig = plt.figure(figsize=(3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color='#383838')
```

#### **LINE WIDTHS**

```python
fig = plt.figure(figsize=(3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color='#00C752', linewidth = 10)
```

```python
# alpha = how transparent the line is
fig = plt.figure(figsize=(3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color='b', alpha=0.99)
```

```python
# Linestyle - dotted
fig = plt.figure(figsize= (3 ,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color='red',linestyle="dotted")
```

```python
fig = plt.figure(figsize= (3 ,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color='red',linestyle="dashed")
```

```python
## MARKER
fig = plt.figure(figsize= (3 ,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color='red',linestyle="dashed", marker="o")
```

```python
fig = plt.figure(figsize= (5 ,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color='red',linewidth=0.2,linestyle="dashed", marker="+", markersize=20)
```

```python
fig = plt.figure(figsize= (5 ,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color='red',linewidth=0.2,linestyle="dashed", marker="o",
        markersize=20, markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green")
```

## **PLOT RANGE**

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

# Setting x and y range from 0 to 1
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])

ax.plot(x, y, color='red',linestyle="dashed")
```
